```python
all_images=os.listdir("images")
#all_images=os.listdir("all-dogs/")

print(f"Total images : {len(all_images)}")

breeds = glob.glob('Annotation/*')
annotation=[]
for b in breeds:
    #print(b)
    annotation+=glob.glob(b+"/*")
    #annotation+=glob.glob(b)
    #print(annotation)
print(f"Total annotation : {len(annotation)}")

breed_map={}
for annot in annotation:
    breed=annot.split("/")[-2]
    index=breed.split("-")[0]
    breed_map.setdefault(index,breed)
    
print(f"Total Breeds : {len(breed_map)}")
```

```python
def bounding_box(image):
    #bpath=root_annots+str(breed_map[image.split("_")[0]])+"/"+str(image.split(".")[0])
    #print(bpath)
    
    bpath="annotations/"+str(image.split(".")[0])+".xml"
    
    tree = ET.parse(bpath)
    root = tree.getroot()
    objects = root.findall("object/[name='undamaged']")
    
    print(len(objects))
    print(objects)
    
    dim = list()
    
    for o in objects:
        bndbox = o.find('bndbox') # reading bound box
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)
        
        dim.append((xmin,ymin,xmax,ymax))
        
        
    #return (xmin,ymin,xmax,ymax)
    return dim
```
